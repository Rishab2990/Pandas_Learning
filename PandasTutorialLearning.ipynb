```python
# MERGING DATAFRAME | INNER , OUTER , LIFT , RIGHT JOINS
```

```python
df1 = pd.DataFrame({
    'id' : [1,2,3,4,5,6],
    'name' : ['Rishab', 'Singh', 'Machine Learning Engineer', 'Rishab Singh Company name is NovaMind', 'This is a  AI  AGI and ASI company', 'Rishab Singh Developed AGI']
})

df2 = pd.DataFrame({
    'id' : [3,4,5,6,7,8],
    'name' : ['AGI' , 'ASI' , 'AI' , 'Rishab' , 'Singh' , 'Rishab Singh is Machine Learning Engineer']
})

innerResult = pd.merge(df1 , df2 , how='inner' , on='id')
innerResult
```

```python
outerResult = pd.merge(df1 , df2 , how='outer' , on='id')
outerResult
```

```python
leftResult = pd.merge(df1 , df2 , how='left' , on='id')
leftResult
```

```python
rightResult = pd.merge(df1 , df2 , how='right' , on='id')
rightResult
```

```python
df1 = pd.DataFrame({
    'id1' : [1,2,3,4,5,6],
    'name' : ['Rishab', 'Singh', 'Machine Learning Engineer', 'Rishab Singh Company name is NovaMind', 'This is a  AI  AGI and ASI company', 'Rishab Singh Developed AGI']
})

df2 = pd.DataFrame({
    'id2' : [3,4,5,6,7,8],
    'name' : ['AGI' , 'ASI' , 'AI' , 'Rishab' , 'Singh' , 'Rishab Singh is Machine Learning Engineer']
})

innerResult = pd.merge(df1 , df2 , how='inner' , left_on='id1' , right_on='id2')
innerResult
```

```python
# CONCATENATING DATAFRAMES 

df1 = pd.DataFrame({
    'id1' : [1,2,3,4,5,6],
    'name' : ['Rishab', 'Singh', 'Machine Learning Engineer', 'Rishab Singh Company name is NovaMind', 'This is a  AI  AGI and ASI company', 'Rishab Singh Developed AGI']
})

df2 = pd.DataFrame({
    'id2' : [3,4,5,6,7,8],
    'name' : ['AGI' , 'ASI' , 'AI' , 'Rishab' , 'Singh' , 'Rishab Singh is Machine Learning Engineer']
})

innerResult = pd.concat([df1,df2],axis =0 , ignore_index=True)
innerResult
```

```python
innerResult2 = pd.concat([df1,df2],axis =0 , keys=['df1','df2'])
innerResult2
```
